```python
import urllib3
import certifi
import re
from bs4 import BeautifulSoup
import pandas as pd
from tabulate import tabulate

http = urllib3.PoolManager(ca_certs=certifi.where())
#assign the website name in a variable
year = 2019
url = f"https://www.metacritic.com/browse/movies/score/metascore/year/filtered?year_selected={year}&sort=desc&view=detailed"

#access the website
r = http.request('GET', f"{url}", headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.94 Safari/537.36'})

#assign result of the website in a variable as string
datastring = str(r.data, "utf-8")

#use BeautifulSout to make the string web result more readable
soup = BeautifulSoup(datastring)

#assign all regex in a variable
pattern1 = r'<h3>(?P<Title>.*)</h3><\/a>'
pattern2 = r'<span>(?P<ReleaseDate>.*\d)</span>'
pattern3 = r'<div class="metascore_w large movie positive">(.*)<\/div>\s+<\/a>\n\s+<\/div>\n\n'
pattern4 = r'><img src="[^/](.*)" alt='
pattern5 = r'<div class="summary">\s+(.*)\s+|\t+</div>'

#assigning result of findall in a variable
idlist1 = re.findall(pattern1, datastring)
idlist2 = re.findall(pattern2, datastring)
idlist3 = re.findall(pattern3, datastring)
idlist4 = re.findall(pattern4, datastring)
idlist5 = re.findall(pattern5, datastring)
idlist6 = idlist5.pop()




#Creating a dictionary from each result to convert it to a pandas dataframe for easy formatting
dict_idlist = {'Title':idlist1, 'ReleaseDate':idlist2, 'MetaScore': idlist3, 'Thumbnail':idlist4, 'Description':idlist5}

#Convert dictianary to datafram
df_idlist = pd.DataFrame(dict_idlist)

#Concatenating h in Thumbnail column because regex removed the h
df_idlist['Thumbnail'] ='h' + df_idlist['Thumbnail'].astype(str)

#Formating dataframe as a tabular view for better viewing

print(tabulate(df_idlist,headers='keys',tablefmt='psql'))
print(f"Found {len(df_idlist)} movies for year {year}.")
```
